```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")

# Display basic information and first few rows
data.info()
print(data.head())



# Distribution plots for numerical variables
for column in data.select_dtypes(include=['number']).columns:
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    sns.histplot(data[column], kde=True, bins=30, ax=axes[0])
    axes[0].set_title(f"Distribution of {column}")
    
    sns.boxplot(x=data[column], ax=axes[1])
    axes[1].set_title(f"Boxplot of {column}")
    
    sns.violinplot(x=data[column], ax=axes[2])
    axes[2].set_title(f"Violin Plot of {column}")
    
    plt.tight_layout()
    plt.show()

# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()

# Pairplot to visualize relationships
sns.pairplot(data, height=2.5)
plt.show()

# Scatter plots for highly correlated variables
correlation_matrix = data.corr().abs()
high_corr = (correlation_matrix > 0.6) & (correlation_matrix < 1.0)

for i in range(len(high_corr.columns)):
    for j in range(i):
        if high_corr.iloc[i, j]:
            plt.figure(figsize=(8, 5))
            sns.scatterplot(x=data.iloc[:, j], y=data.iloc[:, i])
            plt.xlabel(data.columns[j])
            plt.ylabel(data.columns[i])
            plt.title(f"Scatter Plot: {data.columns[j]} vs {data.columns[i]}")
            plt.show()

# Boxen plot for better visualization of distributions
for column in data.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 5))
    sns.boxenplot(x=data[column])
    plt.title(f"Boxen Plot of {column}")
    plt.show()

# Swarm plot for understanding distributions along with scatter points
for column in data.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 5))
    sns.swarmplot(x=data[column])
    plt.title(f"Swarm Plot of {column}")
    plt.show()

# Jointplot to show relationship with density estimation
for column in data.select_dtypes(include=['number']).columns[:2]:
    sns.jointplot(x=data[column], y=data[data.columns[-1]], kind="kde", height=6)
    plt.show()
```
